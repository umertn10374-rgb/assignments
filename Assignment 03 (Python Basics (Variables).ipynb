3. Write a program in which use all the operators we can use in Python

```python

print("--- Arithmetic Operators ---")
a, b = 15, 4
print(f"Addition ({a} + {b}):", a + b)
print(f"Subtraction ({a} - {b}):", a - b)
print(f"Multiplication ({a} * {b}):", a * b)
print(f"Division ({a} / {b}):", a / b)
print(f"Floor Division ({a} // {b}):", a // b)  
print(f"Modulus/Remainder ({a} % {b}):", a % b)
print(f"Exponentiation ({a} ** {b}):", a ** b) 

print("\n--- Assignment Operators ---")
x = 10  
print("Initial x:", x)
x += 5  # Same as x = x + 5
print("After x += 5:", x)
x *= 2  # Same as x = x * 2
print("After x *= 2:", x)


print("\n--- Comparison Operators ---")
m, n = 10, 20
print(f"Is {m} equal to {n}? (==):", m == n)
print(f"Is {m} not equal to {n}? (!=):", m != n)
print(f"Is {m} greater than {n}? (>):", m > n)
print(f"Is {m} less than or equal to {n}? (<=):", m <= n)

print("\n--- Logical Operators ---")
p, q = True, False
print(f"p and q:", p and q)  # True only if both are True
print(f"p or q:", p or q)    # True if at least one is True
print(f"not p:", not p)      # Reverses the boolean value

print("\n--- Identity Operators ---")
list1 = [1, 2, 3]
list2 = [1, 2, 3]
list3 = list1

print("list1 is list3:", list1 is list3)      # True (They point to the same object)
print("list1 is list2:", list1 is list2)      # False (Same content, different objects)
print("list1 is not list2:", list1 is not list2) # True

print("\n--- Membership Operators ---")
fruits = ["apple", "banana", "cherry"]
print("Is 'banana' in fruits?:", "banana" in fruits)
print("Is 'orange' not in fruits?:", "orange" not in fruits)

print("\n--- Bitwise Operators ---")
# 5 in binary: 0101
# 3 in binary: 0011
g, h = 5, 3
print(f"Bitwise AND ({g} & {h}):", g & h)   # 0001 (Decimal 1)
print(f"Bitwise OR ({g} | {h}):", g | h)    # 0111 (Decimal 7)
print(f"Bitwise XOR ({g} ^ {h}):", g ^ h)   # 0110 (Decimal 6)
print(f"Bitwise NOT (~{g}):", ~g)           # -(g+1)
print(f"Bitwise Left Shift ({g} << 1):", g << 1) # Shifts bits left (multiplies by 2)
```

4. Completes the following steps of small task:
Mention Marks of English , Islamiat and Maths out of 100 in 3 different variables
Mention Variable of Total Marks and assign 300 to it
Calculate Percentage


```python

english_marks = 85
islamiat_marks = 90
maths_marks = 95



total_marks = 300


obtained_marks = english_marks + islamiat_marks + maths_marks
percentage = (obtained_marks / total_marks) * 100

# Printing the results
print(f"English Marks: {english_marks}")
print(f"Islamiat Marks: {islamiat_marks}")
print(f"Maths Marks: {maths_marks}")
print(f"Total Marks: {total_marks}")
print(f"Percentage Obtained: {percentage:.2f}%")
```
